```python
from transformers import pipeline

pipe = pipeline("translation", model="facebook/nllb-200-distilled-600M")
```

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("facebook/nllb-200-distilled-600M")
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/nllb-200-distilled-600M")
```

```python
translator = pipeline('translation', model=model, tokenizer=tokenizer, src_lang="eng_Latn", tgt_lang='hin_Deva', max_length = 400)
```

```python
import re
import random
eng_file = 'English.txt'
hin_file = 'Hindi.txt'
# Defining lines as a list of each line
with open(eng_file, 'r', encoding='utf-8') as f:
  lines = f.read().strip().split('\n')
with open(hin_file, 'r', encoding='utf-8') as f:
  lines2 = f.read().strip().split('\n')

lines = [" ".join(re.findall(r"[A-Za-z0-9]+",line)) for line in lines]
lines2 = [re.sub(r"%s|\(|\)|<|>|%|[a-z]|[A-Z]|_",'',line) for line in lines2]

# Grouping lines by response pair
pairs = list(zip(lines,lines2))
random.shuffle(pairs)
```

```python
for pair in pairs[:10]:
    print(f'English Text: {pair[0]}')
    print(f'Actual Hindi Text: {pair[1]}\n')
```
